```python
features = ['Age','Sex','Job','Housing','Saving accounts','Checking account','Credit amount','Duration']
```

```python
target = 'Risk'
```

```python
df_model = df[features + [target]].copy()
```

```python
df_model.head()
```

```python
from sklearn.preprocessing import LabelEncoder
import joblib
```

```python
cat_cols = df_model.select_dtypes(include='object').columns.drop('Risk')
```

```python
le_dict = {}
```

```python
cat_cols
```

```python
for col in cat_cols:
    le = LabelEncoder()
    df_model[col] = le.fit_transform(df_model[col])
    le_dict[col] = le
    joblib.dump(le, f'{col}_encoder.pkl')
```

```python
le_target = LabelEncoder()
```

```python
target
```

```python
df_model[target] = le_target.fit_transform(df_model[target])
```

```python
df_model[target].value_counts()
```

```python
joblib.dump(le_target, 'target_encoder.pkl')
```

```python
df_model.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_model.drop(target, axis=1)
```

```python
y = df_model[target]
```

```python
df["Saving accounts"].unique()
```

```python
df["Checking account"].unique()
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1, stratify=y)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
def train_model(model,param_grid, x_train, y_train, x_test, y_test):
    gird = GridSearchCV(model,param_grid, cv= 5, scoring="accuracy", n_jobs= -1)
    gird.fit(x_train, y_train)
    best_model = gird.best_estimator_
    y_pred = best_model.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    return best_model, acc, gird.best_params_
   
```

```python
dt = DecisionTreeClassifier(random_state=1, class_weight='balanced')
dt_params_grid = {
    "max_depth": [3, 5, 7, 10, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
```

```python
best_dt, acc_dt, params_dt = train_model(dt, dt_params_grid, x_train, y_train, x_test, y_test)
```

```python
print("Decision Tree Accuracy:", acc_dt)
```

```python
print("Best Decision Tree Model Parameters:", params_dt)
```

```python
rf = RandomForestClassifier(random_state=1, class_weight='balanced', n_jobs=1)
```

```python
rf_params_grid = {
    "n_estimators": [100, 200],
    "max_depth": [5, 7, 10, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
```

```python
best_rf, acc_rf, params_rf = train_model(rf, rf_params_grid, x_train, y_train, x_test, y_test)
```

```python
print("Radom Forest Accuracy:", acc_rf)
```

```python
print("Best Parameters:", params_dt)
```

```python
et = ExtraTreesClassifier(random_state=1, class_weight='balanced', n_jobs=-1)
```

```python
et_params_grid = {
    "n_estimators": [ 100, 200],
    "max_depth": [5, 7, 10, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
```

```python
best_et, acc_et, params_et = train_model(et, et_params_grid, x_train, y_train, x_test, y_test)
```

```python
print("Extra trees Accuracy:", acc_et)
```

```python
print("Best Parameters:", params_et)
```

```python
xgb = XGBClassifier(random_state=1, scale_pos_weight= (y_train == 0).sum()/(y_train == 1).sum(), use_label_encoder=False, eval_metric='logloss')
```

```python
xgb_params_grid = {
    "n_estimators": [100, 200],
    "max_depth": [3, 5, 7],
    "learning_rate": [0.01, 0.1, 0.2],
    "subsample": [0.7, 1.0]
}
```

```python
best_xgb, acc_xgb, params_xgb = train_model(xgb, xgb_params_grid, x_train, y_train, x_test, y_test)
```

```python
print("XGB Accuracy:", acc_xgb)
```

```python
print("Best Parameters:", params_xgb)
```
